## 0. Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.applications import VGG16
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.utils import img_to_array
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
```

## 1. Load VGG16 model

```python
def load_model(input_shape):
    model = VGG16(weights="imagenet",
                  include_top=False,
                  input_shape=input_shape,
                  pooling='avg')
    return model
```

## 2. Preprocessing with a CNN and predicting with an RNN - Option A

### 2.1. Create a CNN preprocessor

```python
def CNN_preprocessor():
    model = load_model(input_shape = (150,150,3))
    return model
```

### 2.2. Create an RNN model

```python
def load_RNN_model():

    RNN1_layer = layers.SimpleRNN(units=10, activation='tanh', return_sequences=True, input_shape = (10,512))
    RNN2_layer = layers.SimpleRNN(units=5, activation='tanh', return_sequences=False)
    classification_layer = layers.Dense(20, activation='softmax')

    model = models.Sequential([
    RNN1_layer,
    RNN2_layer,
    classification_layer
    ])

    return model
```

```python
def build_model():
    model = load_RNN_model()

    opt = optimizers.Adam(learning_rate=1e-4)
    model.compile(loss='categorical_crossentropy',
                  optimizer=opt,
                  metrics=['accuracy'])
    return model
```

## 3. Add RNN layers to the CNN model (VGG16) - Option B

### 3. 1. Layers of the VGG16

```python
print(len(vgg16_model.layers))
vgg16_model.layers
```

### 3. 2. Set the original layers to non-trainable

```python
def set_nontrainable_layers(model):
    '''Set the layers as non-trainable'''
    model.trainable = False
    return model
```

### 3.3. Add last layers and RNN model

```python
def add_last_layers(model):
    '''
    1. Load set_nontrainable function
    2. Add trainable Flatten an Dense layers
    '''
    base_model = set_nontrainable_layers(model)
    dense_layer = layers.Dense(10, activation='relu')

    reshape_layer = layers.Reshape((-1, 10))
    RNN1_layer = layers.SimpleRNN(units=10, activation='tanh', return_sequences=True)
    RNN2_layer = layers.SimpleRNN(units=5, activation='tanh', return_sequences=True)
    classification_layer = layers.Dense(20, activation='softmax')

    model = models.Sequential([
    base_model,
    flatten_layer,
    dense_layer,
    reshape_layer,
    RNN1_layer,
    RNN2_layer,
    classification_layer
    ])

    return model
```
